### Preferred Methods

```python
import plotly.graph_objects as go
import pandas as pd


# Create a copy of the relevant columns
df_mini = df[['native_language', 'foreign_language', 'proficiency']].copy()
df_mini['foreign_language'].fillna('', inplace=True)

# Splitting and stripping whitespaces from 'native_language'
df_n_exploded = df_mini['native_language'].str.split(',', expand=True).stack().str.strip().reset_index(level=1, drop=True)

# Splitting and stripping whitespaces from 'foreign_language'
df_exploded = df_mini['foreign_language'].str.split(',', expand=True).stack().str.strip().reset_index(level=1, drop=True)

# Create the nodes and links for the Sankey diagram
sources = []
targets = []
values = []

# Sort unique native languages, learning languages, and proficiency levels
native_languages = sorted(df_n_exploded.unique())
learning_languages = sorted(df_exploded.unique())
proficiency_levels = sorted(df_mini['proficiency'].unique())

# Create mappings
native_map = {lang: i for i, lang in enumerate(native_languages)}
learn_map = {lang: i + len(native_languages) for i, lang in enumerate(learning_languages)}
prof_map = {level: i + len(native_languages) + len(learning_languages) for i, level in enumerate(proficiency_levels)}

# Populate the sources, targets, and values
for idx, row in df_mini.iterrows():
    native_list = row['native_language'].split(',')
    native_list = [lang.strip() for lang in native_list]  # Strip whitespaces
    learning_list = row['foreign_language'].split(',')
    learning_list = [lang.strip() for lang in learning_list]  # Strip whitespaces
    for native_lang in native_list:
        native_idx = native_map[native_lang]
        for lang in learning_list:
            learning_idx = learn_map[lang]
            prof_idx = prof_map[row['proficiency']]
            sources.extend([native_idx, learning_idx])
            targets.extend([learning_idx, prof_idx])
            values.extend([1, 1])

# Define the Sankey diagram
fig = go.Figure(go.Sankey(
    node=dict(
        pad=15,
        thickness=20,
        line=dict(color="black", width=0.5),
        label=list(native_map.keys()) + list(learn_map.keys()) + list(prof_map.keys())
    ),
    link=dict(
        source=sources,
        target=targets,
        value=values
    )
))

fig.update_layout(title_text="Flow of Native Languages to Languages Being Learned to Proficiency Levels", font_size=10)

# Show the figure
fig.show()

# Save the figure as a PNG file
fig.write_image("sankey_diagram.png")
```

## What do language learners prefer/prioritize?

```python
from matplotlib.ticker import MaxNLocator

# Features to plot
features = {
    'f_cost': 'Cost',
    'f_time': 'Time',
    'f_prestige': 'Prestige',
    'f_entertainment': 'Entertainment',
    'f_convenience': 'Convenience',
    'f_flexibility': 'Flexibility'
}

# Calculate value counts for each feature
value_counts = {features[feature]: df[feature].value_counts().sort_index() for feature in features}

# Create a DataFrame from the value counts
value_counts_df = pd.DataFrame(value_counts).fillna(0).astype(int)

# Plot each feature as a separate bar plot
fig, axs = plt.subplots(nrows=3, ncols=2, figsize=(13, 10))
axs = axs.flatten()

for i, (feature, counts) in enumerate(value_counts_df.items()):
    ax = axs[i]
    counts.plot(kind='bar', ax=ax, color='#98d1d1')
    
    # Set plot labels and title
    ax.set_ylabel('Number of Responses', fontsize=12)
    ax.set_xlabel('Ranking (1 = Most Important, 6 = Least Important)', fontsize=12)

    ax.set_title(f'{feature.capitalize()} Preference', fontsize=15)  # Bold title
    
    # Remove ticks
    ax.tick_params(axis='both', which='both', length=0)  # Remove ticks
    
    # Rotate x-axis labels
    ax.set_xticks(range(len(counts.index)))
    ax.set_xticklabels(counts.index, rotation=0, ha='center', va='center', fontsize= 12)
    
    # Increase space between label and x-axis spine
    ax.tick_params(axis='x', pad=10)
    
    # Add the number count on top of each bar
    for p in ax.patches:
        ax.annotate(str(p.get_height()), (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 5), textcoords='offset points')

    # Remove spines
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)

    # Format y-axis as integer
    ax.yaxis.set_major_formatter('{:,.0f}'.format)
    
    # Set y-axis integer ticks
    ax.yaxis.set_major_locator(MaxNLocator(integer=True))

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

### What is the most commone study methods?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is your DataFrame

df_effective_exploded = df['most_effective'].str.split(',', expand=True).stack().str.strip().reset_index(level=1, drop=True)
df_other_exploded = df['most_effective_other'].str.split(',', expand=True).stack().str.strip().reset_index(level=1, drop=True)

combined = pd.concat([df_effective_exploded, df_other_exploded])
combined = combined[combined != 'Other:']

vals = combined.value_counts()

# Sort the value counts from high to low
value_counts_sorted = vals.sort_values(ascending=True)

# Round the counts to integers
value_counts_int = value_counts_sorted.astype(int)
# Define categories to highlight
highlight_categories = ['Designated Course', 'Duolingo']

# Determine colors for the bars
colors = ['#98d1d1' if category not in highlight_categories else '#54bebe' for category in value_counts_int.index]

# Plot the horizontal bar chart
fig, ax = plt.subplots(figsize=(12, 8))
value_counts_int.plot(kind='barh', color=colors, ax=ax, width=0.7)

# Improve the plot aesthetics
ax.set_xlabel('Number of Responses', fontsize=14)
ax.set_ylabel('Effectiveness Categories', fontsize=14)
ax.set_title('What Do Language Learners Find the Most Effective?', fontsize=16)
ax.grid(True, which='both', linestyle='--', linewidth=0.5, color='gray', axis='x')

# Remove the lines on the x and y axes
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)

# Remove the y-axis ticks but keep the labels
ax.yaxis.set_ticks_position('none')

# Set x-axis labels to be integers
ax.xaxis.set_major_locator(plt.MaxNLocator(integer=True))

# Add counts on each bar as annotations
for index, value in enumerate(value_counts_int):
    ax.text(value, index, str(int(value)), va='center', ha='right', color='black', fontsize=12)

# Adjust y-axis labels for better readability
ax.set_yticklabels(ax.get_yticklabels(), fontsize=12)

plt.tight_layout()
plt.show()
```

### budget vs most effective method of study


```python

# Create the DataFrame
analysis_a =  df[['budget', 'most_effective', 'most_effective_other']]

# Splitting comma-separated strings into lists
analysis_a['most_effective'] = analysis_a['most_effective'].str.split(',')
analysis_a['most_effective_other'] = analysis_a['most_effective_other'].str.split(',')

analysis_a.dropna(subset=['most_effective', 'most_effective_other'], inplace=True)

# Combining 'most_effective' and 'most_effective_other' columns into a single list
analysis_a['effective_methods'] = analysis_a.apply(lambda row: row['most_effective'] + row['most_effective_other'] if row['most_effective_other'] else row['most_effective'], axis=1)

# Explode the 'effective_methods' column
analysis_a = analysis_a.explode('effective_methods')

# Drop the original columns
analysis_a.drop(columns=['most_effective', 'most_effective_other'], inplace=True)

analysis_a = analysis_a[analysis_a.effective_methods != 'Other:']
analysis_a
```

```python
# Sort the DataFrame by budget in descending order
sorted_budget = analysis_a.sort_values(by='budget', ascending=False)

# Plotting
plt.figure(figsize=(10, 6))
plt.barh(sorted_budget['effective_methods'], sorted_budget['budget'], color='skyblue')
plt.xlabel('Budget (in USD)')
plt.ylabel('Effective Methods')
plt.title('Budget Allocation for Effective Study Methods')
plt.gca().invert_yaxis()  # Invert y-axis to have the highest budget at the top
plt.show()
```
